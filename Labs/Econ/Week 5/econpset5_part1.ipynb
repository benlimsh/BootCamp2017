The steady-state equations are as follows:

$$ \overline{c} = (1 - \tau) [\overline{w} + (\overline{r} - \delta) \overline{k}] + \overline{k} + \overline{T} - \overline{k} $$

$$\overline{c}^{-\gamma} = \beta E_t \{ \overline{c}^{-\gamma} [(\overline{r} - \delta)(1-\tau) + 1 ]\} $$

$$ 0 = \overline{c}^{-\gamma} \overline{w} (1 - \tau) $$

$$r_t = \alpha \overline{K}^{\alpha - 1} (e^{z_t}) ^ {1 - \alpha}$$

$$\overline{w} = 0$$

$$ \tau [\overline{w}\mathscr{l}_t + (\overline{r} - \delta) \overline{k}] = \overline{T} $$

$$ z_t = \overline{z} $$

We can use $r_t = \alpha \overline{K}^{\alpha - 1} (e^{z_t}) ^ {1 - \alpha}$ and $\overline{c}^{-\gamma} = \beta E_t \{ \overline{c}^{-\gamma} [(\overline{r} - \delta)(1-\tau) + 1 ]\} $ to solve for the steady-state value of capital. Substituting, we get $\overline{k} = [\frac{1}{\alpha}(\frac{1}{e^{\overline{z}}})^{1 - \alpha} (\frac{1 - \beta}{\beta(1 - \tau)} + \delta)]^{\frac{1}{\alpha - 1}} =  [\frac{1}{\alpha}(e^{\overline{z}})^{\alpha - 1} (\frac{1 - \beta}{\beta(1 - \tau)} + \delta)]^{\frac{1}{\alpha - 1}} = e^{\overline{z}} [ (\frac{1 - \beta}{\beta(1 - \tau)} + \delta) \frac{1}{\alpha}]^{\frac{1}{\alpha -1}}$. By substituting the parameters in, this expression evaluates to $\overline{K} = 7.287$, and $F = 2.213$.





```python
import numpy as np
from scipy.optimize import fsolve
import math

gamma = 2.5
beta = 0.98
alpha = 0.4
delta = 0.1
z = 0
tau = 0.05
l = 1

def ex5_numerical(p):
    
    k, c, w, r, T = p
    eqn21 = c - (1-tau)*(w + (r-delta)*k) - T
    eqn25 = beta*((r-delta)*(1-tau)+1)-1
    eqn26 = w*(1-tau)/(c**gamma)
    eqn27 = (alpha*(k**(alpha-1))*(np.exp(z))**(1-alpha))-r
    eqn28 = w
    
    return (eqn21, eqn25, eqn26, eqn27, eqn28)

k, c, w, r, T = fsolve(ex5_numerical, (1,1,1,1,1))
O = (k**alpha)*(l*(np.exp(z)))**(1-alpha)

ex5results = '''
kbar = %s
cbar = %s
wbar = %s
rbar = %s
Tbar = %s
Investment = %s
Output = %s
'''
print(ex5results % (k, c, w, r, T, k, O))
```

```python
gamma6 = 2.5
psi6 = 1.5
beta6 = 0.98
alpha6 = 0.4
a6 = 0.5
delta6 = 0.1
z6 = 0
tau6 = 0.05

def ex6_numerical(p):
    k6, c6, w6, r6, T6, l6 = p
    eqn21 = c6 - (1-tau6)*(w6*l6 + (r6-delta6)*k6) - T6
    eqn25 = (c6**-gamma6) - beta6*((c6**-gamma6)*((r6-delta6)*(1-tau6)+1))
    eqn26 = a6/((1-l6)**psi6) - (c6**-gamma6)*w6*(1-tau6)
    eqn27 = alpha6*(k6**(alpha6-1))*((l6*np.exp(z6))**(1-alpha6)) - r6
    eqn28 = (1-alpha6)*(k6**alpha6)*((l6*np.exp(z6))**(-alpha6))*np.exp(z6) - w6
    eqn29 = tau6*(w6*l6 + (r6-delta6)*k6) - T6
    
    return (eqn21, eqn25, eqn26, eqn27, eqn28, eqn29)

k6, c6, w6, r6, T6, l6 = fsolve(ex6_numerical, (0.5,0.5,0.5,0.5,0.5,0.5))
O6 = (k6**alpha6)*(l6*(np.exp(z6)))**(1-alpha6)


ex6results = '''
kbar = %s
cbar = %s
wbar = %s
rbar = %s
Tbar = %s
Investment = %s
Output = %s
l = %s
'''
print(ex6results % (k6, c6, w6, r6, T6, k6, O6, l6))
```

```python
import numpy as np
import matplotlib.pyplot as plt
import time

def rouwen(rho, mu, step, num):
    '''
    Adapted from Lu Zhang and Karen Kopecky. Python by Ben Tengelsen.
    Construct transition probability matrix for discretizing an AR(1)
    process. This procedure is from Rouwenhorst (1995), which works
    well for very persistent processes.
    INPUTS:
    rho  - persistence (close to one)
    mu   - mean and the middle point of the discrete state space
    step - step size of the even-spaced grid
    num  - number of grid points on the discretized process
    OUTPUT:
    dscSp  - discrete state space (num by 1 vector)
    transP - transition probability matrix over the grid
    '''

    # discrete state space
    dscSp = np.linspace(mu -(num-1)/2*step, mu +(num-1)/2*step, num).T

    # transition probability matrix
    q = p = (rho + 1)/2.
    transP = np.array([[p**2, p*(1-q), (1-q)**2], \
                    [2*p*(1-p), p*q+(1-p)*(1-q), 2*q*(1-q)], \
                    [(1-p)**2, (1-p)*q, q**2]]).T


    while transP.shape[0] <= num - 1:

        # see Rouwenhorst 1995
        len_P = transP.shape[0]
        transP = p * np.vstack((np.hstack((transP, np.zeros((len_P, 1)))), np.zeros((1, len_P+1)))) \
                + (1 - p) * np.vstack((np.hstack((np.zeros((len_P, 1)), transP)), np.zeros((1, len_P+1)))) \
                + (1 - q) * np.vstack((np.zeros((1, len_P+1)), np.hstack((transP, np.zeros((len_P, 1)))))) \
                + q * np.vstack((np.zeros((1, len_P+1)), np.hstack((np.zeros((len_P, 1)), transP))))

        transP[1:-1] /= 2.


    # ensure columns sum to 1
    if np.max(np.abs(np.sum(transP, axis=1) - np.ones(transP.shape))) >= 1e-12:
        print('Problem in rouwen routine!')
        return None
    else:
        return transP.T, dscSp

#transP.T is the transition matrix
#dscSp is the vector grid for z

alpha = 0.35
beta = 0.98
rho = 0.9
sigma = 0.02


mu = 0
num = 25
minz = -5*sigma
maxz = 5*sigma
step =  (maxz - minz)/(num-1)
A = alpha * beta
kbar = A**(1/(1-alpha))
pi, zgrid = rouwen(rho, mu, step, num) #transP.T 
kgrid = np.linspace(0.5*kbar, 1.5*kbar, 25)
V = np.zeros((25,25))

print(zgrid)
print(kgrid)
```

```python
e = np.zeros((25,25,25)) #N = 9
for h in range(25): #loop over z
    for i in range(25): #loop over k
        for j in range(25): #loop over k'
                 e[h, i, j] = np.log(np.exp(zgrid[h])*(kgrid[i]**alpha) - kgrid[j])
print(e[:,0,0])
```

```python
VFtol = 1e-6
VFdist = 7.0
VFmaxiter = 3000
Vmat = np.zeros((25,25,25))  # initialize Vmat matrix, which is now 3D (z,k,k')
#Vstore = np.zeros((25, VFmaxiter))  # initialize Vstore array
VFiter = 1
start_time = time.clock()
V = np.zeros((25,25))


while VFdist > VFtol and VFiter < VFmaxiter:
    
    TV = np.copy(V)
    V = pi.T @ V
    for h in range(25):  #loop over z'
        for i in range(25):  # loop over k
            for j in range(25):  # loop over k'
                Vmat[h, i, j] = e[h, i, j] + beta * V[h, j] 
                 
    #Vstore[:, :, VFiter] = V.reshape(25,25)  # store value function at each
    # iteration for graphing later
    V = Vmat.max(axis=2)  # apply max operator to Vmat (to get V(k))
    VFdist = (np.absolute(V - TV)).max()  # check distance between value
    # function for this iteration and value function from past iteration
    VFiter += 1

H = np.argmax(Vmat, axis=2)  # find the index of the optimal k'
VFI_time = time.clock() - start_time
if VFiter < VFmaxiter:
    print('Value function converged after this many iterations:', VFiter)
else:
    print('Value function did not converge')
print('VFI took ', VFI_time, ' seconds to solve')


VF = V  # solution to the functional equation
```
